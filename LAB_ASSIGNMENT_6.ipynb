Question 1: Word Count - Count the frequency of each word
Input:

• hadoop is fast

• hadoop is scalable

Part A: Manual MapReduce (Python Implementation)

Step 1: Define Input Data

```python
data = [
    "hadoop is fast",
    "hadoop is scalable"
]
```

Step 2: Mapper Phase

```python
mapped = []

for line in data:
    words = line.split()
    for word in words:
        mapped.append((word, 1))

print(mapped)
```

Step 3: Shuffle and Sort Phase

```python
shuffle = {}

for word, count in mapped:
    if word not in shuffle:
        shuffle[word] = []
    shuffle[word].append(count)

print(shuffle)
```

Step 4: Reducer Phase

```python
result = {}

for word, counts in shuffle.items():
    result[word] = sum(counts)

print(result)
```

Question 3: Average Word Length (Per Word) - Compute the average length of each word.

Input: data science data big data

Part A: Manual MapReduce (Python Implementation)

Step 1: Define Input Data

```python
data = ["data science data big data"]
```

Step 2: Mapper Phase

```python
mapped = []

for line in data:
    words = line.split()
    for word in words:
        mapped.append((word, len(word)))

print(mapped)
```

Step 3: Shuffle and Sort Phase

```python
shuffle = {}

for word, length in mapped:
    if word not in shuffle:
        shuffle[word] = []
    shuffle[word].append(length)

print(shuffle)
```

Step 4: Reducer Phase

```python
result = {}

for word, lengths in shuffle.items():
    avg = sum(lengths) / len(lengths)
    result[word] = avg

print(result)
```

Part B: MapReduce using mrjob Framework

Step 1: Create Python File

```python
%%writefile avg_word_length.py
from mrjob.job import MRJob

class AvgWordLength(MRJob):

    def mapper(self, _, line):
        for word in line.split():
            yield word, (len(word), 1)

    def reducer(self, key, values):
        total_length = 0
        total_count = 0

        for length, count in values:
            total_length += length
            total_count += count

        yield key, total_length / total_count

if __name__ == '__main__':
    AvgWordLength.run()
```

```python
%%writefile top5.py
from mrjob.job import MRJob

class Top5Words(MRJob):

    def mapper(self, _, line):
        for word in line.split():
            yield word.lower(), 1

    def reducer(self, key, values):
        yield key, sum(values)

    def reducer_final(self):
        top5 = sorted(self.items(), key=lambda x: x[1], reverse=True)[:5]
        for word, count in top5:
            yield word, count

if __name__ == '__main__':
    Top5Words.run()
```

Question 6: Compute average marks for each student.

Input:
A 80
B 70
A 90
B 60
A 100

Part A: Manual MapReduce

```python
data = [
    "A 80",
    "B 70",
    "A 90",
    "B 60",
    "A 100"
]

# Mapper
mapped = []
for line in data:
    student, marks = line.split()
    mapped.append((student, int(marks)))

# Shuffle
shuffle = {}
for student, marks in mapped:
    if student not in shuffle:
        shuffle[student] = []
    shuffle[student].append(marks)

# Reducer
result = {}
for student, marks in shuffle.items():
    result[student] = sum(marks) / len(marks)

print(result)
```

Part B: Using mrjob

```python
%%writefile avg_marks.py
from mrjob.job import MRJob

class AvgMarks(MRJob):

    def mapper(self, _, line):
        student, marks = line.split()
        yield student, (int(marks), 1)

    def reducer(self, key, values):
        total = 0
        count = 0
        for marks, c in values:
            total += marks
            count += c
        yield key, total / count

if __name__ == '__main__':
    AvgMarks.run()
```

Question 7: Compute average salary per department and Highest Paid Department (Based on Average
Salary)

Input:
HR 50000
IT 70000
HR 60000
IT 80000

Part A: Manual

```python
data = [
    "HR 50000",
    "IT 70000",
    "HR 60000",
    "IT 80000"
]

mapped = []
for line in data:
    dept, salary = line.split()
    mapped.append((dept, int(salary)))

shuffle = {}
for dept, salary in mapped:
    if dept not in shuffle:
        shuffle[dept] = []
    shuffle[dept].append(salary)

result = {}
for dept, salaries in shuffle.items():
    result[dept] = sum(salaries) / len(salaries)

print(result)

# Highest paid department
highest = max(result, key=result.get)
print("Highest Paid Department:", highest)
```

Part B: mrjob

```python
%%writefile avg_salary.py
from mrjob.job import MRJob

class AvgSalary(MRJob):

    def mapper(self, _, line):
        dept, salary = line.split()
        yield dept, (int(salary), 1)

    def reducer(self, key, values):
        total = 0
        count = 0

        for value in values:
            salary, c = value
            total += salary
            count += c

        yield key, total / count

if __name__ == '__main__':
    AvgSalary.run()
```

Question 8: Computer average temperature per country

New York,38

London,29

Tokyo,35

New York,32

Delhi,45

Ambala,35

Part A: Manual MapReduce

```python
data = [
    "New York,38",
    "London,29",
    "Tokyo,35",
    "New York,32",
    "Delhi,45",
    "Ambala,35"
]

# Mapper
mapped = []
for line in data:
    city, temp = line.split(",")
    mapped.append((city, int(temp)))

# Shuffle
shuffle = {}
for city, temp in mapped:
    if city not in shuffle:
        shuffle[city] = []
    shuffle[city].append(temp)

# Reducer
result = {}
for city, temps in shuffle.items():
    result[city] = sum(temps) / len(temps)

print(result)
```

Part B: Using mrjob

```python
%%writefile avg_temp.py
from mrjob.job import MRJob

class AvgTemp(MRJob):

    def mapper(self, _, line):
        city, temp = line.split(",")
        yield city, (int(temp), 1)

    def reducer(self, key, values):
        total = 0
        count = 0

        for value in values:
            temp, c = value
            total += temp
            count += c

        yield key, total / count

if __name__ == '__main__':
    AvgTemp.run()
```

Question 9: Redo Q8 with the dataset available on:
https://www.kaggle.com/datasets/heemalichaudhari/global-land-temperatures

Part A: Manual MapReduce

```python
mapped = []

with open("global.csv") as f:
    next(f)  # skip header
    for line in f:
        parts = line.strip().split(",")

        try:
            temp = parts[1]
            if temp == "":
                continue

            temp = float(temp)
            city = parts[3]

            mapped.append((city, temp))
        except:
            pass

# Shuffle
shuffle = {}
for city, temp in mapped:
    shuffle.setdefault(city, []).append(temp)

# Reducer
result = {}
for city, temps in shuffle.items():
    result[city] = sum(temps) / len(temps)

# Print few results
for i, (k, v) in enumerate(result.items()):
    print(k, v)
    if i == 9:
        break
```

Part B: Using mrjob

```python
%%writefile avg_temp_kaggle.py
from mrjob.job import MRJob

class AvgTemp(MRJob):

    def mapper(self, _, line):
        parts = line.split(",")

        # Skip header
        if parts[0] == "dt":
            return

        try:
            temp = parts[1]
            if temp == "":
                return

            temp = float(temp)
            city = parts[3]   # ✅ correct column

            yield city, (temp, 1)

        except:
            pass

    def reducer(self, key, values):
        total = 0
        count = 0

        for temp, c in values:
            total += temp
            count += c

        yield key, total / count

if __name__ == '__main__':
    AvgTemp.run()
```
