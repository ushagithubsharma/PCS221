Part I: Randomized Sales Data Generation (NumPy)

```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Set random seed (use your roll number here)
np.random.seed(42)

# Step 3: Generate sales data (12 months × 4 categories)
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# Step 4: Create DataFrame
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

df = pd.DataFrame(sales_data, index=months, columns=columns)

print("DataFrame:\n", df)
```

Part II: Data Manipulation and Analysis(Pandas)

```python
# 1. First 5 rows and summary stats
print("\nFirst 5 rows:\n", df.head())
print("\nSummary:\n", df.describe())
```

```python
# 2. Total sales per category & per month
category_total = df.sum()
monthly_total = df.sum(axis=1)

print("\nCategory Total:\n", category_total)
print("\nMonthly Total:\n", monthly_total)
```

```python
# 3. Average growth between consecutive months
growth = df.pct_change()
avg_growth = growth.mean()

print("\nAverage Growth:\n", avg_growth)
```

```python
# 4. Add new columns
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

print("\nUpdated DataFrame:\n", df)
```

```python
# 5. Apply discount (change roll number accordingly)
roll_number = 42  # CHANGE THIS

if roll_number % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.9
else:
    df["Clothing"] = df["Clothing"] * 0.85

print("\nAfter Discount:\n", df)
```

Part III: Visualization

```python
# Line Plot
df[["Electronics", "Clothing", "Home & Kitchen", "Sports"]].plot(marker='o')
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
# Box Plot
plt.figure()
sns.boxplot(data=df[["Electronics", "Clothing", "Home & Kitchen", "Sports"]])
plt.title("Sales Distribution")
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value      
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row.      
iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python
import numpy as np

arr = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Absolute value
print("Absolute:\n", np.abs(arr))
```

```python
# ii. Percentiles
flat = arr.flatten()

print("25th percentile:", np.percentile(flat, 25))
print("50th percentile:", np.percentile(flat, 50))
print("75th percentile:", np.percentile(flat, 75))

print("Column-wise percentiles:\n", np.percentile(arr, [25,50,75], axis=0))
print("Row-wise percentiles:\n", np.percentile(arr, [25,50,75], axis=1))
```

```python
# iii. Mean, Median, Std

print("Flattened Mean:", np.mean(flat))
print("Flattened Median:", np.median(flat))
print("Flattened Std:", np.std(flat))

print("\nColumn-wise Mean:", np.mean(arr, axis=0))
print("Row-wise Mean:", np.mean(arr, axis=1))
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).    
i. Find floor, ceiling and truncated value,rounded values

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Truncate:", np.trunc(a))
print("Round:", np.round(a))
```

Q.4 Write a program to swap two elementsin a list in Python Using a Temporary Variable.

```python
lst = [10, 20, 30, 40]

pos1 = 1
pos2 = 3

temp = lst[pos1]
lst[pos1] = lst[pos2]
lst[pos2] = temp

print("After swap:", lst)
```

Q.5 Write a Python program to swap two elementsin a set by converting it to a list first.

```python
s = {10, 20, 30, 40}

lst = list(s)

pos1 = 0
pos2 = 2

lst[pos1], lst[pos2] = lst[pos2], lst[pos1]

s = set(lst)

print("After swap:", s)
```
