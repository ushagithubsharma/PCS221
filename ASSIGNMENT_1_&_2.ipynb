1. Generate data file using random module (20000 × 10000)


```python
import random

rows = 20000
cols = 10000

with open("data.txt", "w") as f:
    for _ in range(rows):
        row = [str(random.randint(0, 100)) for _ in range(cols)]
        f.write(" ".join(row) + "\n")

print("Data file created successfully!")
```

2. Assign subsequent rows to first row elements

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

first_row = matrix[0]

for i in range(1, len(matrix)):
    matrix[i] = first_row.copy()

print("Updated Matrix:")
for row in matrix:
    print(row)
```

3. Group similar elements into matrix

```python
from collections import defaultdict

arr = [1, 2, 3, 1, 2, 3, 4, 4]

group = defaultdict(list)

for num in arr:
    group[num].append(num)

matrix = list(group.values())

print("Matrix of grouped elements:")
for row in matrix:
    print(row)
```

4. Add and subtract matrices

```python
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]

rows = len(A)
cols = len(A[0])

# Addition
add = [[A[i][j] + B[i][j] for j in range(cols)] for i in range(rows)]

# Subtraction
sub = [[A[i][j] - B[i][j] for j in range(cols)] for i in range(rows)]

print("Addition:")
for row in add:
    print(row)

print("\nSubtraction:")
for row in sub:
    print(row)
```

5. n×n matrix where each sub-matrix opposite corners sum is even

```python
n = 3

matrix = [[(i + j) for j in range(n)] for i in range(n)]

valid = True

for i in range(n - 1):
    for j in range(n - 1):
        if (matrix[i][j] + matrix[i+1][j+1]) % 2 != 0:
            valid = False

print("Matrix:")
for row in matrix:
    print(row)

if valid:
    print("Condition satisfied: Opposite corner sums are even")
else:
    print("Condition NOT satisfied")
```

6. Row-wise element addition in tuple matrix

```python
tuple_matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

result = []

for row in tuple_matrix:
    result.append(sum(row))

print("Row-wise addition result:")
print(result)
```
